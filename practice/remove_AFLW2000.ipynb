```python
list_2000 = '/Data/nruiz9/data/facial_landmarks/AFLW2000/filename_list_filtered.txt'
list_normal = '/Data/nruiz9/data/facial_landmarks/AFLW/aflw_cropped_loose/filename_list_filtered.txt'
```

```python
fid = open(list_2000, 'r')
dict_2000 = dict()
for line in fid:
    line = line.strip('\n').split('/')
    dict_2000[line[-1]] = 1
print len(dict_2000)
fid.close()
```

```python
fid = open(list_normal, 'r')
naked = list_normal.strip('.txt')
train = open(naked + '_noAFLW2000_train.txt', 'wb')
test = open(naked + '_noAFLW2000_test.txt', 'wb')
test_dict = dict()
train_dict = dict()
for line in fid:
    line = line.strip('\n')
    name = line.split('/')[-1]
    if name in dict_2000.keys():
        test.write(line + '\n')
        test_dict[line] = 1
    else:
        train.write(line + '\n')
        train_dict[line] = 1

print len(test_dict), len(train_dict)
```
